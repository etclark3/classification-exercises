```python
def titanic_data():
    return pd.read_sql('SELECT * FROM passengers', get_connection('titanic_db'))
```

```python
def get_titanic_data():
    filename = "titanic.csv"
    
    # if file is available locally, read it
    if os.path.isfile(filename):
        return pd.read_csv(filename)
    
    # if file not available, acquire data from SQL database and write it as csv locally for future use
    else:
        # read the SQL query into a dataframe
        titanic = titanic_data()
        
        # Write that dataframe to disk for later. Called "caching" the data for later.
        titanic.to_csv(filename)

        return titanic
```

    alpha = 0.05
    
    H0 = col+' and species are independent'
    
    Ha = 'There is a relationship between species and '+col
    
    observed = pd.crosstab(tel_train.churn_Yes, tel_train[col])
    
    chi2, p, degf, expected = stats.chi2_contingency(observed)
    
    if p < alpha:

```python
def tel_cat_test(tel_train, col):
    alpha = 0.05
    H0 = col+' and churn are independent'
    Ha = 'There is a relationship between churn and '+col
    observed = pd.crosstab(tel_train.churn_Yes, tel_train[col])
    chi2, p, degf, expected = stats.chi2_contingency(observed)
    if p < alpha:
        print('We reject that', H0)
        print(Ha)
    else:
        print('We fail to reject that', H0)
        print('There appears to be no relationship between churn and '+col)
```

```python
def telco_analysis(tel_train, col):
    tel_vis(tel_train, col)
    tel_cat_test(tel_train, col)
```
