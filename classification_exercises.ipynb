```python
import env
```

```python
def titanic_data():
    return pd.read_sql('SELECT * FROM passengers', get_connection('titanic_db'))
```

```python
def get_titanic_data():
    filename = "titanic.csv"
    
    # if file is available locally, read it
    if os.path.isfile(filename):
        return pd.read_csv(filename)
    
    # if file not available, acquire data from SQL database and write it as csv locally for future use
    else:
        # read the SQL query into a dataframe
        titanic = titanic_data()
        
        # Write that dataframe to disk for later. Called "caching" the data for later.
        titanic.to_csv(filename)

        return titanic
```

Make a function named `get_iris_data` that returns the data from the `iris_db` on the codeup data science database as a pandas data frame. The returned data frame should include the actual name of the species in addition to the `species_id`s. Obtain your data from the _Codeup Data Science Database_. 

```python
def get_iris_data():
    return pd.read_sql('SELECT * FROM measurements join species using(species_id)', get_connection('iris_db'))
```

Make a function named `get_telco_data` that returns the data from the `telco_churn` database in SQL. In your SQL, be sure to join all 4 tables together, so that the resulting dataframe contains all the contract, payment, and internet service options. Obtain your data from the _Codeup Data Science Database_. 

```python
def get_telco_data():
    return pd.read_sql('SELECT * FROM customers join contract_types using(contract_type_id) join payment_types using(payment_type_id) join internet_service_types using(internet_service_type_id)', get_connection('telco_churn'))
```

```python
def grab_iris_data():
    filename = "iris.csv"
    
    if os.path.isfile(filename):
        return pd.read_csv(filename)
    
    else:
        iris = get_iris_data()
        iris.to_csv(filename)
        return iris
    
grab_iris_data().head()
```

```python
def grab_telco_data():
    filename = "telco.csv"
    
    if os.path.isfile(filename):
        return pd.read_csv(filename)
    
    else:
        telco = get_telco_data()
        telco.to_csv(filename)
        return telco
```
